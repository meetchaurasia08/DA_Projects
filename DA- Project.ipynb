<h1> Project - Pandas and Data visualisation</h1>

```python
import pandas as pd
data = "https://raw.githubusercontent.com/swapnilsaurav/Dataset/refs/heads/master/Iris.csv"
df = pd.read_csv(data)
print(df)
```

```python
print("Size of the dataset: ", df.shape)
print("Top 5 Rows: \n",df.head(5))
print("Bottom 5 Rows: \n",df.tail(5))
```

```python
print("Different columns and their datatypes:\n",df.info())
```

```python
print("Major Statistics:\n",df.describe())
```

```python
print("Checking for missing values:\n",df.isnull().sum())
```

```python
print("Checking for unique values in Species column")
df_species = df.drop_duplicates(subset="Species")
print(df_species)
```

```python
#checking if the dataset is balanced or not
print("Number of rows:\n",df.value_counts("Species"))
```

<h2> Visualisations </h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns

#plot1: count plot
sns.countplot(x="Species",data=df)
plt.title("Count Plot")
plt.show()
```

```python
#plot2: Scatter plot for 2 numeric values
sns.scatterplot(x="SepalLengthCm",y="PetalLengthCm",
                hue="Species",data=df)
plt.show()
```

```python
## Plot 3: HISTOGRAM
fig,axes = plt.subplots(2,2,figsize=(10,10))
axes[0,0].set_title("SepalLengthCm")
axes[0,0].hist(df["SepalLengthCm"],bins=12)

axes[0,1].set_title("SepalWidthCm")
axes[0,1].hist(df["SepalWidthCm"],bins=10)

axes[1,0].set_title("PetalLengthCm")
axes[1,0].hist(df["PetalLengthCm"],bins=10)

axes[1,1].set_title("PetalWidthCm")
axes[1,1].hist(df["PetalWidthCm"],bins=10)
plt.show()
```

```python
# Plot 4: Boxplot
sns.boxplot(x="Species",y="PetalLengthCm",data=df)
plt.show()
```
